## Removing NULL values

```python
raw_data_count = len(recent_grads.index)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads.index)
print(raw_data_count)
print(cleaned_data_count)
```

## Employing Pandas Scatter Matrix Plot

## Scatter Matrix of 3 variables wil create a 3*3 matrix

```python
my_colors = ['r','g','b','k','y','m','c']
```

**Bar Plots in Pandas**
## Comparison of % of women from 10 highest and lowest paying Majors

```python
recent_grads[:10].plot.bar(x='Major', y='Women', legend=False,color = my_colors)
recent_grads[163:].plot.bar(x='Major', y='Women', legend=False, color = my_colors)
```

# Comparison of Unemployment Rate from 10 highest and lowest paying majors

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False,color = my_colors)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False,color = my_colors)
```
